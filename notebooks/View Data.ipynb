```python
from torch.utils.data import DataLoader
import util
import torch

from torchvision import transforms
img_transform = transforms.Compose([
    transforms.ToTensor(),
])
    
data_train = util.NotSoCLEVRDataset(csv_file='../data/data_train.csv',
                                    img_file='../data/data_train.h5',
                                    transform=img_transform)
```

```python
import matplotlib.pyplot as plt
% matplotlib inline

fig = plt.figure(figsize=(10, 15))
for i in range(4,10):
    img = data_train[i]['image'].permute(1, 2, 0).numpy()
    fig.add_subplot(3, 2, i-3)
    plt.imshow(img / 255)
    plt.title('{} \n({}) {}'.format(data_train[i]['task'], 
          'non-relational' if data_train[i]['type'] else 'relational', 
          data_train[i]['answer']))
    plt.axis('off')
    plt.tight_layout()
plt.show()
```
